# **Objetos Index**: Indexação, Seleção e Filtragem

Vamos ver um pouco mais sobre indexação, como selecionar (acessar) valores e/ou colunas e a filtrar.

Para fazer esses processos, seguimos a anatomia listada no quadro abaixo.


|Operação                                    | Sintaxe             | Resultado    |
|:---------:                                 |:-------------------:|:------------:|
|Seleciona uma coluna                        | ``df[col]``         | ``Series``   |
|Seleciona duas colunas                      | ``df[[col1, col2]]``| ``DataFrame``|
|Seleciona uma linha usando rótulo           | ``df.loc[rotulo]``  | ``Series``   |
|Seleciona uma linha usando inteiro          | ``df.iloc[i]``      | ``Series``   |
|Recorta o dataframe da linha 2 a 5          | ``df[2:5]``        |``DataFrame`` |
|Seleciona linhas usando um vetor booleano   | ``df[vetor_bool]``    | ``DataFrame``|

Para vermos como essas operações se aplicam, vamos usar os mesmos dados usado na aula anterior. Vamos definir de novo Series e Daframe a partir de um dicionário.

```
import pandas as pd
import numpy as np
```

```
popBrasilDicionario = {'Amazonas': 3938336, 'São Paulo': 44396484, 'Bahia': 15203934, 'Minas Gerais': 20869101, 'Roraima': 505665, 'Distrito Federal': 2914830}
areaBrasilDicionario = {'Amazonas': 1571825, 'São Paulo': 248219, 'Bahia': 564760, 'Minas Gerais': 869101, 'Roraima': 45665, 'Distrito Federal': 4830}
```

Criando uma Series:

```
populacaoSeries = pd.Series(popBrasilDicionario)
```

```
areaSeries = pd.Series(areaBrasilDicionario)
```

Criando um DataFrame:

```
estadosBrasil = pd.DataFrame( {'População': populacaoSeries, 'Área': areaSeries})
estadosBrasil
```

Vou só adicionar mais uma coluna, a de Densidade.

```
estadosBrasil['Densidade'] = estadosBrasil['População'] / estadosBrasil['Área']
estadosBrasil
```

### Selecionando uma coluna

O retorno vai ser uma Series.

```
estadosBrasil['População']
```

```
estadosBrasil['Área']
```

```
# é possível acessar só com . e o nome da coluna
# ATENÇÃO!!!!
# Se o nome da coluna for composto (formado por mais de uma palavra), esse método NÃO vai funcionar

estadosBrasil.Área
```

## Selecionando duas colunas



O retorno vai ser um DataFrame.

```
estadosBrasil[['População', 'Área']] # sempre usar 2 colchetes concêntricos
```

### Selecionando duas colunas e invertendo a ordem



```
estadosBrasil[['Área', 'População']]
```

## .loc: Selecionando uma linha usando o rótulo/'nome'

O retorno vai ser uma Series.

```
estadosBrasil.loc['Minas Gerais']
```

Retornou em notação científica.

## Selecionando uma linha usando o índice

Retorna uma Series, assim como o .loc.

|índice                                    | Coluna 0             | Coluna 1    |
|:---------:                                 |:-------------------:|:------------:|
|Índice 1                        | 00         | 01   |
|Índice 2                      | 10 | 11 |
|Índice 3          | 20  | 21   |

No nosso exemplo:

|Estado     | População | Área    | Densidade |
|:---------: |:-------------------:|:------------:| :------------:|
|Amazonas      | 00   | 01   |  |
|São Paulo     | 10 | 11 |  |
|Bahia     | 20  | 21   |  |
|Minas     | 20  | 21   |  |
|Roraima     | 20  | 21   |  |
|Distrito Federal     | 20  | 21   |  |

```
# vai puxar todos os dados da 4ª linha

estadosBrasil.iloc[3]
```

## 'Fatiando' o dataframe a partir de alguns nomes: Slicing

```
estadosBrasil['Bahia': 'Roraima']
```

### Filtrando o DataFrame com vetor booleano

O retorno será um DataFrame.

Aqui o objetivo é definir uma avaliação booleana e, a partir dela, retornar um vetor ou conjunto de valores booleanos. O objetivo não é só retornar um booleano com esses valores, mas surge como uma possibilidade de usar esse caminho para obter outros resultados, como o exemplo abaixo.

```
vetor_booleano = estadosBrasil['Área'] > 50000
vetor_booleano
```

Obtivemos o valor tipo bool no vetor_booleano.

Mas o objetivo é poder usar isso de outras forma, como no exemplo abaixo em que obtemos um DataFrame a partir de um bool.

```
estadosBrasil[vetor_booleano]
```

## Transpondo o dataframe

Basta adicionar .T

```
estadosBrasil.T
```

Note que nesse exemplo e em outros obtivemos os números em notação científica.

Para evitar isso, podemos usar o seguinte comando:


```
pd.options.display.float_format = '{:.2f}'.format
```

```
estadosBrasil.T
```

## Acessando os valores do DataFrame usando ``values``

Vai retornar um array.

```
estadosBrasil.values
```

Para acessar apenas os valores da linha 0:

```
estadosBrasil.values[0]
```

```
estadosBrasil.values[1]
```

## Acessando os valores do DataFrame usando ``loc`` com condições

```
estadosBrasil.loc[estadosBrasil['Densidade'] < 10 ]
```

```
# Aqui não retorna a linha, mas apenas a Densidade

estadosBrasil.loc[estadosBrasil['Densidade'] < 10, 'Densidade' ]
```

### Substituindo um valor

Agora que selecionamos os estados em que a Densidade é menor que 10, podemos substituir o valor da densidade por outro, por exemplo, por 90000.

```
estadosBrasil.loc[estadosBrasil['Densidade'] < 10, 'Densidade'] = 90000
```

Verificando como está agoira:

```
estadosBrasil
```

## Acessando os valores do DataFrame usando ``iloc``

```
estadosBrasil
```

```
estadosBrasil.iloc[2]
```

Acessando só a Densidade desse estado.

```
estadosBrasil.iloc[2,2]
```

Agora se quisermos trocar o valor da Densidade desse estado, seguimos como abaixo.

```
estadosBrasil.iloc[2,2] = 100000
```

```
estadosBrasil
```

# Introdução às estruturas de dados do Pandas: **Series e DataFrames**

Esses dois são os tipos de estruturas de dados mais usados na ciência de dados. São formas de estruturar dados, a partir dos quais é possível realizar algumas operações.

Veremos como organizar/arranjar/distribuir/escrever/definir listas, dicionários, etc. nos formatos de Series e DataFrame e, em seguida, veremos como operá-las.

## Series

Uma Series é um array com rótulos de 1 dimensão e é capaz de conter qualquer tipo de dado do Python. Os rótulos são chamados de índices (```index```).

O método mais básico de criar uma série no pandas é definindo:



>>```s = pd.Series(data=parametro_com_nome_do_conjunto_de_dados, index = indice)```





O parâmetro ```dados``` pode ser de diferentes tipo:



*   Lista
*   Dicionário
*   ND-Array
*   Valor escalar




O índice é um rótulo/nome para cada linha.

Carregando o Pandas e o NumPy para ilustramos esse conceito para alguns parâmetros diferentes.

````
import pandas as pd
import numpy as np
````

### Definindo uma ***Series*** a partir de uma lista

Definindo uma lista a ser usada: a lista de renda por classe social, baseada em dados do IBGE de 2020.

````
# Seja a lista abaixo

rendas = [20900, 10450, 4180, 2090, 0]
rendas
````

Definindo a Series:

````
classeSocialSemIndex = pd.Series(data = rendas)
classeSocialSemIndex
````

Note que passamos apenas o primeiro parâmetro, o *data*, de modo que só foi criada uma coluna com os dados e a coluna mais à esquerda tem números (índices) automáticos.

Agora caso queiramos da rótulos para cada linha de uma forma particular, vamos precisar passas os índices. Para indicar ínices de uma forma desejada, podemos **passar o segundo parâmetro**:

````
indices = ['A', 'B', 'C', 'D', 'E']

# já criando a Series com índice

classeSocialComIndex = pd.Series(data = rendas, index = indices)
classeSocialComIndex
````

#### Acessando os valores de uma Series

Acessando valores de uma Series sem index

````
classeSocialSemIndex[0]
````

````
classeSocialSemIndex[2]
````

````
classeSocialSemIndex[len(rendas)-1]
````

Acessando valores de uma Series com index: acesso via rótulo

````
classeSocialComIndex['B']
````

````
classeSocialComIndex['E']
````

Uma forma de fazer uma consulta rápida dos índices é:

````
classeSocialComIndex.index
````

Uma forma de acessa em partes:

````
classeSocialComIndex['B':'D']
````

````
classeSocialSemIndex[0:len(rendas)]
````

Podemos ainda mudar os índices das Series. Vamos trocar os valores da lista indices:

````
# indices = ['A', 'B', 'C', 'D', 'E']

indices = [10, 8, 32, 54, 178]

classeSocialComIndex = pd.Series(data = rendas, index = indices)
classeSocialComIndex
````

Acessando os valores:

````
classeSocialComIndex[10]
````

Podemos fazer essa mesma troca de índices sem precisar criar a Series novemente, mas, em vez disso, só reatribuindo os índices como segue (para ilustrar, vou trocar só os índices novamente):

````
indices = [47, 8, 131, 1,0.5]

#classeSocialComIndex.index = indices
````

````
classeSocialComIndex.index = indices
classeSocialComIndex
````

````
# acessando

classeSocialComIndex[47]
````

## DataFrame

É uma estrutura de dados com rótulos de 2 dimensões, podendo ter colunas de diferentes tipos. É comparável às tabelas do excel ou de banco de dados. É a estrutura do Pandas mais usada.

### Definindo um ***DataFrame*** a partir de uma lista

````
# definindo a lista

lista_df = [2, 4, 6, 8, 10]
````

````
# definindo o DataFrame
# cada coluna é como uma Series

pd.DataFrame(lista_df)
````

````
# renomenado a coluna

pd.DataFrame(lista_df, columns = ['Pares'])
````

### Definindo um ***DataFrame*** a partir de um dicionário

````
# definindo o df

pd.DataFrame({'Ímpares': [1, 3, 5, 7, 9]})
# o nome da coluna é uma chave; o valor é uma lista
````

### Definindo um ***DataFrame*** a partir de um dicionário com 2 listas





````
pd.DataFrame( { 'pares': [2, 4, 6, 8, 10],
               'ímpares': [1, 3, 5, 7, 9]
}
)
````

### Definindo um ***DataFrame*** a partir de uma Series do Pandas

Vou criar uma tabela com a população e área de alguns estados brasileiros.

````
popBrasilDicionario = {'Amazonas': 3938336, 'São Paulo': 44396484, 'Bahia': 15203934}
areaBrasilDicionario = {'Amazonas': 1571825, 'São Paulo': 248219, 'Bahia': 564760}
````

````
popBrasilDicionario
````

````
areaBrasilDicionario
````

Criando as 2 Series com esses dicionários:

````
populacaoSeries = pd.Series(popBrasilDicionario)
populacaoSeries
````

````
areaSeries = pd.Series(areaBrasilDicionario)
areaSeries
````

  Criando o DataFrame, para isso, vou passar um dicionário para o Dataframe sendo que a chave é o cabeçalho da coluna e o valor sá as Series que criei com os valores de População e Área:

````
estadosBrasil = pd.DataFrame( {'População': populacaoSeries, 'Área': areaSeries})
estadosBrasil
````

## Operações com DataFrame

### Acessando elementos: ```.loc``` e ```.iloc```

É possível acessar de 2 formas;



*   .loc - acessa por rótulo (nome do índice)
*   .iloc - acessa por posição numérica (0, 1, 2, ...)



As contagens dos índices são na seguinte ordem:



*   0: População
*   1: Área


````
## acessando por .loc - rótulo/nome do índice
estadosBrasil.loc['São Paulo', 'Área']
````

Aqui é preciso ter cuidado, porque a matriz de dados é do tipo 3x2, de modo que:


*   00 é a Pop. de AM e 01 é a Área de AM;
*   10 é a Pop. de SP e 11 é a Área de SP;
*   20 é a Pop. de BA e 21 é a Área de BA.

````
### acessarndo por .iloc- por posição numérica
estadosBrasil.iloc[1, 1]
````

### Acessando colunas

````
estadosBrasil['População']
````

````
estadosBrasil['Área']
````

### Soma vertical

Caso queiramos somar os valores de uma coluna:

````
estadosBrasil['Área'].sum()
````

### Criando uma nova coluna

````
estadosBrasil['Densidade'] = estadosBrasil['População'] / estadosBrasil['Área']
estadosBrasil
````

### Soma horizontal (só por curiosidade)

Uma soma dos elementos da linha aqui não faz sentido porque estaremos somando população com área. Vou mostrar aqui só para mostrar a possibilidade.

Para fazer a soma, basta adicionar o parâmetro axis=1, pois axis=0 indica a soma por coluna (padrão).

````
estadosBrasil.sum(axis=1).astype(int) # adicionei .astype(int) para não ficar em notação científica
````

Outra maneira de escrever essa soma para não ficar em notação científica:

````
pd.options.display.float_format = '{:.0f}'.format
estadosBrasil.sum(axis=1)
````

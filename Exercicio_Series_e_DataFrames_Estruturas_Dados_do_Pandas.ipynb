# Transformando Array em DF |  4) Transforme o ndarray abaixo em um dataframe (esse exercício é lindo)


O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

O método ```.random.nornal()``` do NumPy é usado para criar uma array (matriz) contendo números aleatórios baseados em uma **distribuição normal** (ou Gaussiana).

Os parâmetros representam:


*   loc: média ("centro") da distribuição | aceita float ou array de float.
*   scale: desvio padrão (width ou spread) da distribuição. Deve ser não-negativa | aceita float ou array de float.
*   size: forma do output | opcional, aceita tupla ou tupla de inteiros.


````python
arr = np.random.normal(100, 10, (20,3))
# primeiro parâmetro: a média de números gerados é ~100, nesse caso, os valores ficam todos na faixa de 80~120
# segundo parâmetro: desvio padrão de 10 (os números variam ~10 unidades em torno de 100)
#terceiro parâmetro: matriz de 20 linhas e 3 colunas
arr
````

Vou gerar um gráfico abaixo só para ficar mais fácil de entender sobre o que está sendo produzido com o comando da array acima:

````python
import matplotlib.pyplot as plt

plt.hist(arr[:, 0], bins=10, alpha=0.5, label='Coluna 0')
plt.hist(arr[:, 1], bins=10, alpha=0.5, label='Coluna 1')
plt.hist(arr[:, 2], bins=10, alpha=0.5, label='Coluna 2')
plt.legend()
plt.title('Distribuição normal (média=100, desvio=10)')
plt.show()
````

Não foi possível quer a distribuição no formato de sino com tão poucas amostras.

Por quesões didáticas, vou aumentar o número de linhas (amostras) só para que a distribuição em forma de sino fique mais evidente:

````python
arr_grande = np.random.normal(100, 10, (10000, 3))

plt.hist(arr_grande[:, 0], bins=50, alpha=0.5, label='Coluna 0')
plt.hist(arr_grande[:, 1], bins=50, alpha=0.5, label='Coluna 1')
plt.hist(arr_grande[:, 2], bins=50, alpha=0.5, label='Coluna 2')
plt.legend()
plt.title('Distribuição normal com 10.000 amostras')
plt.show()
````

Agora vou escrever esse array como DataFrame.

Num primeiro momento, podemos pensar em escrever o ```arr```em formato de Series e, depois, passar para o DataFrame como um dicionário, como abaixo:



```
# Transformando em Series

arrayTesteSeries = pd.Series(arr)
arrayTesteSeries
```

```
# Transformando em DataFrame

arrayTesteDF = pd.DataFrame( {'Distribuição': arrayTesteSeries } )
arrayTesteDF



Mas isso não vai dar certo porque pd.Series não aceita array 2D.


O arr é uma matriz (20x3), e **uma Series só comporta 1 dimensão**. Ao tentar fazer pd.Series(arr), o pandas não sabe o que fazer com as 3 colunas.
Por isso pd.DataFrame(arr) funcionou diretamente: o DataFrame já é feito para dados 2D.

A lógica de passar por Series funcionaria se o array fosse 1D, por exemplo:



```
# Isso funciona porque é 1D (só uma coluna)
coluna0 = pd.Series(arr[:, 0])
arrayDataFrame = pd.DataFrame({'Coluna0': coluna0})
```



Passando direto para o DataFrame, portanto:

````python
arrayDataFrame = pd.DataFrame(arr)
arrayDataFrame
````

# 5) Inserindo Index num DF | Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

Vou usar a indicação que consta na documentação [pandas.DataFrame.reindex
](https://https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html), que indica a possibilidade de adicionar essa renomeação de índices como parâmetro.

# 6) Criando Nova Coluna | No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

````python
dataFrameRenomeado['media'] = (dataFrameRenomeado['x1'] + dataFrameRenomeado['x3'] + dataFrameRenomeado['x3'])/3
dataFrameRenomeado
````

Uma outra forma de fazer, seria usarndo a função ```.mean()```do Python, que já calcula a média de um conjunto de números. Mas, para isso, é recomendado fazer uma cópia do ```dataFrameRenomeado``` e trabalhar a aplicação desse método em cima da cópia.

````python
dfMediaHorizontal = dataFrameRenomeado.copy()
dfMediaHorizontal['media'] = dfMediaHorizontal.mean(axis=1) # já que a soma vai ser horizontal
dfMediaHorizontal
````

## Curiosidade: Criando uma nova linha - Operação Vertical

Vamos proceder de forma similar caso queiramos adicionar uma nova linha, porém vamos precisar concatenar essa linha com a tabela original.

Só a título de curiosidade, vou usar o ```.max()```.

````python
dfMaxVertical = dataFrameRenomeado.copy()

# criando a nova linha
novaLinha = pd.DataFrame([dfMaxVertical.max(axis=0)], index = ['máximo valor'])

# concatenando essa nova linha no dfMaxVertical
dfMaxVertical = pd.concat( [dfMaxVertical, novaLinha] )
dfMaxVertical
````
